```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = pd.read_csv('Wholesale customers data.csv')
```

```python
a.head()
```

```python
from scipy.cluster.hierarchy as shc
```

```python
from sklearn.preprocessing import normalize
```

```python
a_scaled=normalize(a)
```

```python
a_scaled=pd.DataFrame(a_scaled, columns=a.columns)
a_scaled.head()
```

```python
from sklearn.preprocessing import normalize
```

```python
import scipy.cluster.hierarchy as shc
```

```python
plt.figure(figsize=(10,7))
plt.title('Dendrograms')
dend=shc.dendrogram(shc.linkage(a_scaled,method='ward'))
```

```python
plt.figure(figsize=(10,7))
plt.title('Dendrograms')
dend=shc.dendrogram(shc.linkage(a_scaled,method='ward'))
plt.axhline(y=6, color='g', linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage= 'ward')
cluster.fit_predict(a_scaled)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(a_scaled['Milk'], a_scaled['Grocery'], c=cluster.labels_)
```
