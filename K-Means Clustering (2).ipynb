```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
X = -2 * np.random.rand(100, 2)

Y = 1 + 2*np.random.rand(50, 2)
```

```python
X[50:100, :] = Y
plt.scatter(X[: , 0], X[ :, 1], s = 50, c ='g')
plt.show()
```

```python
from sklearn.cluster import KMeans
Km = KMeans(n_clusters=2)
Km.fit(X)
```

```python
Km.cluster_centers_
```
