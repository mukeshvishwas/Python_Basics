```python
# Imputing Missing values (mean, median, mode)
data["Total Number of Registered Motor Vehicles (in thousands)"].fillna(value=data["Total Number of Registered Motor Vehicles (in thousands)"].mean(), inplace = True)
data["Road Length (in kms)"].fillna(value=data["Road Length (in kms)"].mean(), inplace = True)

data["Number of Accidents per Lakh Population"].fillna(value=data["Number of Accidents per Lakh Population"].mean(), inplace = True)
data["Number of Accidents per Ten Thousand Vehicles"].fillna(value=data["Number of Accidents per Ten Thousand Vehicles"].mean(), inplace = True)
data["Number of Accidents Per Ten Thousand Kms of Roads"].fillna(value=data["Number of Accidents Per Ten Thousand Kms of Roads"].mean(), inplace = True)

data["Number of Persons Killed Per Lakh Population"].fillna(value=data["Number of Persons Killed Per Lakh Population"].mean(), inplace = True)
data["Number of Persons Killed Per Ten Thousand Vehicles"].fillna(value=data["Number of Persons Killed Per Ten Thousand Vehicles"].mean(), inplace = True)
data["Number of Persons Killed per Ten Thousand Kms of Roads"].fillna(value=data["Number of Persons Killed per Ten Thousand Kms of Roads"].mean(), inplace = True)

data["Number of Persons Injured per Lakh Population"].fillna(value=data["Number of Persons Injured per Lakh Population"].mean(), inplace = True)
data["Number of Persons Injured Per Ten Thousand Vehicles"].fillna(value=data["Number of Persons Injured Per Ten Thousand Vehicles"].mean(), inplace = True)
data["Number of Persons Injured Per Ten Thousand Kms of Roads"].fillna(value=data["Number of Persons Injured Per Ten Thousand Kms of Roads"].mean(), inplace = True)

data.isna().sum()
```

## Step 4- Data Visualization
### Uni-variate Analysis

- Univariate analysis is perhaps the simplest form of statistical analysis. 
- Has the purpose to describe a single variable distribution in one sample
- Like other forms of statistics, it can be inferential or descriptive. 
- The key fact is that only one variable is involved. 
- Univariate analysis explores each variable in a data set, separately. 
- It looks at the range of values, as well as the central tendency of the values
- Univariate analysis can yield misleading results in cases in which multivariate analysis is more appropriate

#### 1. Univariate Scatter Plot

- It plots different observations/values of the same variable corresponding to the index/observation number.

```python
# Q1. Tell us over the years, accidents are increasing or decreasing
plt.xticks(rotation=90)
plt.scatter(data['Years'],data['Total Number of Road Accidents (in numbers)'], c='r')
plt.show()
```

#### Data Insights:
- Over the years number of accidents are increasing

#### 2. Univariate Line Plot

```python
# Q2. Tell us over the years, No of people killed are increasing or decreasing
plt.xticks(rotation=90)
plt.plot(data['Years'],data['Total Number of Persons Killed (in numbers)'], c='b')
plt.show()
```

#### Data Insights:
- Over the years, No of people killed are increasing

#### 3. Histogram (Univariate)

- Histograms are similar to bar charts which display the counts or relative frequencies of values falling in different class intervals or ranges.

```python
# Q3. Show us the Distribution of Number of Registered Motor Vehicles (in thousands)
plt.hist(data['Total Number of Registered Motor Vehicles (in thousands)'])
plt.show()
```

#### Data Insights:
- Over the years, more number of 25K to 50k vehicles registered

### Bi-variate Analysis

#### 1. Bi-Variate Line Plot

```python
# Q4. Show yearwise No of Accidents, No of people Killed, No of People Injured, Population etc
plt.figure(figsize=(12,7))
plt.xticks(rotation=90)
plt.plot(data["Years"],data["Total Number of Road Accidents (in numbers)"],c='r',label="Accidents")
plt.plot(data["Years"],data["Total Number of Persons Killed (in numbers)"],c='b',label="Killed")
plt.plot(data["Years"],data["Total Number of Persons Injured (in numbers)"],c='m',label="Injured")
plt.plot(data["Years"],data["Population of India (in thousands)"],c='g',label="Total Population")
plt.legend()
plt.show()
```

```python
# Q4. Show yearwise No of Accidents, No of people Killed, No of People Injured, Population etc
plt.figure(figsize=(12,7))
plt.xticks(rotation=90)
sns.boxplot(x=data["Years"],y=data["Total Number of Road Accidents (in numbers)"],c='r',label="Accidents")
sns.boxplot(x=data["Years"],y=data["Total Number of Persons Killed (in numbers)"],c='b',label="Killed")
sns.boxplot(x=data["Years"],y=data["Total Number of Persons Injured (in numbers)"],c='m',label="Injured")
sns.boxplot(x=data["Years"],y=data["Population of India (in thousands)"],c='g',label="Total Population")
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(1, 1, figsize=(16,9))
plt.xticks(rotation=90)
sns.boxplot( y="Total Number of Persons Killed (in numbers)", x= "Years", data=data, orient="v")
plt.show()
```
